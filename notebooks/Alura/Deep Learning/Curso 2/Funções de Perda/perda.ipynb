```python
if torch.cuda.is_available():
    device = torch.device("cuda")
else:
    device = torch.device("cpu")

print(device)
```

```python
from sklearn import datasets

wine = datasets.load_wine()
data = wine.data
target = wine.target

print(data.shape)
print(target.shape)

print(wine.feature_names)
print(wine.target_names)

```

```python
data[-1]
```

```python
class WineClassifier(nn.Module):
    def __init__(self, input_size, hidden_size, output_size ):
        super(WineClassifier, self).__init__()
        self.hidden = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.output = nn.Linear(hidden_size, output_size) 
        self.softmax = nn.Softmax()

    def forward(self, x):
        feature = self.relu(self.hidden(x))
        output = self.softmax(self.output(feature))
        return output
```

```python
input_size = data.shape[1]
hidden_size = 32
output_size = len(wine.target_names)

model = WineClassifier(input_size, hidden_size, output_size)
model.to(device)
```

```python
criterion = nn.CrossEntropyLoss().to(device)

Xtns = torch.from_numpy(data).float()
Ytns = torch.from_numpy(target).long()

Xtn = Xtns.to(device)
Ytn = Ytns.to(device)

print(Xtn.dtype)
print(Ytns.dtype)
```

```python
pred = model(Xtn)

print(pred.shape)
print(Ytn.shape)
```

```python
loss = criterion(pred, Ytn)
print(loss)
```
